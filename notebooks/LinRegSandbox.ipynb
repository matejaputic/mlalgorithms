```python
%matplotlib notebook
import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
dataset_filename = 'data/ex1data2.csv'
df = pd.read_csv(dataset_filename, header=None)
#df.apply(lambda x: x * x)
df_n = df.apply(lambda x: x * x)
l2_norms = df_n.sum()
df_n = (df / l2_norms)
df_n
```

```python
threedee = plt.figure().gca(projection='3d')
threedee.scatter(df_n.index, df_n[0], df_n[1], c='r')
plt.show()
```

```python
df_surf = pd.DataFrame()
df_surf[0] = np.ones(50)
df_surf[1] = np.linspace(*threedee.axes.get_xlim())
df_surf[2] = np.linspace(*threedee.axes.get_ylim())
print(df_surf)
theta = [86273.8, 2.03215e+06, -153233]
df_surf.dot(theta)
```
